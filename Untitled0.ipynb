#Klasifikasi Risiko Tinggi Putus Sekolah SD Berdasarkan Jenis Kelamin dan Status Sekolah di Indonesia Tahun 2023

```python
import pandas as pd

# Load dataset
url = "https://raw.githubusercontent.com/dietadestina/datamining_project/refs/heads/main/jumlah-putus-sekolah-menurut-jenis-kelamin-dan-status-sekolah-tiap-provinsi-indonesia-sd-2023.csv"
df = pd.read_csv(url)
df
```

```python
# Cek semua kolom
df.columns
```

```python
df = df.rename(columns={
    'Laki-laki (L) - Negeri': 'Laki_Negeri',
    'Laki-laki (L) - Swasta': 'Laki_Swasta',
    'Laki-laki (L) - Subjml': 'Jml_Laki',
    'Perempuan (P) - Negeri': 'Perempuan_Negeri',
    'Perempuan (P) - Swasta': 'Perempuan_Swasta',
    'Perempuan (P) - Subjml': 'Jml_Perempuan',
    'Laki-laki dan Perempuan - Jumlah': 'Total_Jumlah'
})

df.columns
```

```python
# Cek missing values
df.isnull().sum()
```

```python
df
```

##Jenis Kelamin

```python
#Buat tabel baru
#Tampilkan hanya kolom yang dibutuhkan
do_jk = df[['Provinsi', 'Jml_Laki', 'Jml_Perempuan', 'Total_Jumlah']]
do_jk
```

```python
# Hitung batas berdasarkan persentil (untuk laki-laki)
q_laki = do_jk['Jml_Laki'].quantile([0.33, 0.66])
q_perempuan = do_jk['Jml_Perempuan'].quantile([0.33, 0.66])
q_total = do_jk['Total_Jumlah'].quantile([0.33, 0.66])

print("Laki-laki:\n", q_laki)
print("\nPerempuan:\n", q_perempuan)
print("\nTotal:\n", q_total)
```

```python
# Fungsi klasifikasi berdasarkan kuartil
def label_kuartil(val, batas1, batas2):
    if val <= batas1:
        return 'Rendah'
    elif val <= batas2:
        return 'Sedang'
    else:
        return 'Tinggi'

# Terapkan ke masing-masing kolom
do_jk['Risiko_Laki'] = do_jk['Jml_Laki'].apply(lambda x: label_kuartil(x, q_laki[0.33], q_laki[0.66]))
do_jk['Risiko_Perempuan'] = do_jk['Jml_Perempuan'].apply(lambda x: label_kuartil(x, q_perempuan[0.33], q_perempuan[0.66]))
do_jk['Risiko_Total'] = do_jk['Total_Jumlah'].apply(lambda x: label_kuartil(x, q_total[0.33], q_total[0.66]))
```

```python
# Lihat distribusi label baru
print("Distribusi Risiko Laki-laki:")
print(do_jk['Risiko_Laki'].value_counts())
print("\nDistribusi Risiko Perempuan:")
print(do_jk['Risiko_Perempuan'].value_counts())
print("\nDistribusi Risiko Total:")
print(do_jk['Risiko_Total'].value_counts())
```

```python
do_jk
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode label klasifikasi
le = LabelEncoder()
do_jk['Label_Laki'] = le.fit_transform(do_jk['Risiko_Laki'])
do_jk['Label_Perempuan'] = le.fit_transform(do_jk['Risiko_Perempuan'])

# Cek hasil encoding
print(le.classes_)
do_jk
```

###Laki-laki

```python
from sklearn.model_selection import train_test_split

# Fitur dan target laki-laki
X_laki = do_jk[['Jml_Laki']]
y_laki = do_jk['Label_Laki']

# Split data
X_train_l, X_test_l, y_train_l, y_test_l = train_test_split(
    X_laki, y_laki, test_size=0.2, stratify=y_laki, random_state=42
)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_l, y_train_l)
y_pred_dt = dt.predict(X_test_l)

print("=== Decision Tree - Laki-laki ===")
print(classification_report(y_test_l, y_pred_dt, target_names=le.classes_))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_l, y_train_l)
y_pred_knn = knn.predict(X_test_l)

print("=== KNN - Laki-laki ===")
print(classification_report(y_test_l, y_pred_knn, target_names=le.classes_))
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train_l, y_train_l)
y_pred_nb = nb.predict(X_test_l)

print("=== Naive Bayes - Laki-laki ===")
print(classification_report(y_test_l, y_pred_nb, target_names=le.classes_))
```

###Perempuan

```python
# Fitur dan target Perempuan
X_perempuan = do_jk[['Jml_Perempuan']]
y_perempuan = do_jk['Label_Perempuan']

# Split data
X_train_l, X_test_l, y_train_l, y_test_l = train_test_split(
    X_perempuan, y_perempuan, test_size=0.2, stratify=y_perempuan, random_state=42
)
```

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_l, y_train_l)
y_pred_dt = dt.predict(X_test_l)

print("=== Decision Tree - Perempuan ===")
print(classification_report(y_test_l, y_pred_dt, target_names=le.classes_))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_l, y_train_l)
y_pred_knn = knn.predict(X_test_l)

print("=== KNN - Perempuan ===")
print(classification_report(y_test_l, y_pred_knn, target_names=le.classes_))
```

```python
nb = GaussianNB()
nb.fit(X_train_l, y_train_l)
y_pred_nb = nb.predict(X_test_l)

print("=== Naive Bayes - Perempuan ===")
print(classification_report(y_test_l, y_pred_nb, target_names=le.classes_))
```
